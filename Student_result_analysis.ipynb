```python
# Student Result Analysis Mini Project:

# This project analyzes student data stored in a CSV file.
# Tasks performed:
# 1. Filter students with marks > 85
# 2. Calculate average marks
# 3. Count students by city
# 4. Find the topper
# 5. Filter Jaipur students with marks > 80
```

```python
import csv

with open("students.csv","r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        print(row)
```

```python
# Filter Students
import csv

with open("students.csv","r") as file:
    reader = csv.DictReader(file)

    print("Students with marks > 85:")

    for row in reader:
        if int(row["marks"]) > 85:
            print(row["name"], row["marks"])
```

```python
# Average Marks
import csv

total = 0
count = 0

with open("students.csv","r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        total += int(row["marks"])
        count += 1

print("Average Marks:", total / count)
```

```python
# City Count

import csv

city_count = {}

with open("students.csv","r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        city = row["city"]

        if city in city_count:
            city_count[city] += 1
        else:
            city_count[city] = 1

print("City-wise Count:", city_count)
```

```python
# Topper
import csv

max_marks = 0
topper = ""

with open("students.csv","r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        marks = int(row["marks"])

        if marks > max_marks:
            max_marks = marks
            topper = row["name"]

print("Topper:", topper, "-", max_marks)
```

```python
# Jaipur Filter
import csv

with open("students.csv","r") as file:
    reader = csv.DictReader(file)

    print("Jaipur students with marks > 80:")

    for row in reader:
        if row["city"] == "Jaipur" and int(row["marks"]) > 80:
            print(row["name"])
```
